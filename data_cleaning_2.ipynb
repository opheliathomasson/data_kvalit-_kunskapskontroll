### Data cleaning of the master csv

#### The columns about the book like author, cover doesn't seem relevant for the purpose. columns like wiki, wand, hair-and eye color, image, height, family_members doesn't seem relevant. 

```python
master.shape # check the number of rows and columns in the dataset
```

```python
master.duplicated().sum() # check for duplicate rows in the dataset
```

```python
master.isnull().sum() # check for missing values in each column to identify any potential issues with data completeness or quality.
```

```python
master = master.drop(columns=['hair_color', 'eye_color', 'born', 'died', 'family_members', 'height', 
                              'image', 'marital_status', 'nationality', 'skin_color', 'wands', 'slug', 'wiki',
                              'author', 'cover', 'dedication', 'pages', 'release_date', 
                              'summary']) # drop columns with too many missing values and not relevant for us

master.head()
```

```python
master = master.dropna(subset=['name']) # drop rows with missing values in the "name" column, as this is a key identifier for the characters and is necessary for analysis.
```

```python
master = master.dropna(thresh=5) # drop rows with less than 5 non-missing values, as these rows may not contain enough information to be useful for analysis.
```
