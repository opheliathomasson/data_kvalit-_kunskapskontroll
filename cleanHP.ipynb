```python
# Läs in och slå ihop de fyra Harry Potter-datafilerna
import pandas as pd
from pathlib import Path

base_path = Path('/Users/matildakosir/Documents/data_manager/datakvalitet/HarryPotter')

files = [
    ('Characters.csv', {'sep': ';', 'engine': 'python'}),
    ('Potions.csv', {'sep': ';', 'engine': 'python'}),
    ('Spells.csv', {'sep': ';', 'engine': 'python'}),
    ('harry_potter_master_data.csv', {'sep': ',', 'engine': 'python'})
]

def clean_column_names(columns):
    return (
        columns.str.strip()
               .str.replace('\n', '_', regex=False)
               .str.replace(' ', '_', regex=False)
               .str.replace('[^0-9A-Za-z_]', '', regex=True)
               .str.lower()
    )


def clean_df(df):
    df = df.copy()
    df.columns = clean_column_names(df.columns)
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].replace({'': pd.NA, 'nan': pd.NA})
    return df


dfs = []
for file_name, kwargs in files:
    path = base_path / file_name
    print(f'Loading {file_name} ...')
    df = pd.read_csv(path, **kwargs)
    df = clean_df(df)
    df['source_file'] = file_name
    dfs.append(df)

combined = pd.concat(dfs, ignore_index=True, sort=False)
combined = combined.drop_duplicates().reset_index(drop=True)

print('\nKombinerade data:')
print('Antal rader:', len(combined))
print('Antal kolumner:', len(combined.columns))
print('Kolumner:', list(combined.columns))

output_path = base_path / 'harry_potter_combined_cleaned.csv'
combined.to_csv(output_path, index=False)
print(f'Rensad och sammanslagen fil sparad till: {output_path}')

combined.head(10)
```
